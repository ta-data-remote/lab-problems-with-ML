## Importing all we need

```python
cookies = pd.read_csv("/Users/timeisenmenger/Desktop/Ironhack /Week8/ML-Competition/Modified-DataFrames/Cookies.csv")
cookies_test = pd.read_csv("/Users/timeisenmenger/Desktop/Ironhack /Week8/ML-Competition/Modified-DataFrames/Cookies_test.csv")
```

```python
cookies.head()
```

## Base for Modeling

```python
# Setup X_train, X_test, y_train, y_test
X_train = cookies.drop(columns = 'quality')
X_test = cookies_test.drop(columns = 'quality')
y_train = cookies["quality"]
y_test = cookies_test["quality"]
```

```python
# ScalerModel
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Linear Regression Model

```python
grid = GridSearchCV(estimator = LinearRegression(), param_grid = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}, cv=10)
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
y_pred_test = grid.predict(X_test_scaled)
```

```python
y_pred_test
```
