## Base for Modeling

```python
# Setup X_train, X_test, y_train, y_test
X_train = cookies2.drop(columns = 'quality')
X_test = cookies2_test.drop(columns = 'quality')
y_train = cookies2["quality"]
y_test = cookies2_test["quality"]
```

```python
# ScalerModel
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Linear Regression Model

```python
grid = GridSearchCV(estimator = LinearRegression(), param_grid = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}, cv=10)
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
y_pred_test = grid.predict(X_test_scaled)
y_pred_test
```

```python
prediction_array_dt_cookies = y_pred_test
```
